```python
def create_sequence_id_at_bat_ordered(df, num_pitches=3):
    """
    Create a 'sequence id' column in the DataFrame that contains information about the previous
    `num_pitches` pitches in the at-bat for the same batter, in chronological order. The information 
    includes Count, Pitch type, Location, and Result. If the at-bat has had fewer than `num_pitches`, 
    return 0.

    Args:
    df (DataFrame): The baseball data DataFrame.
    num_pitches (int): Number of previous pitches to include in the sequence.

    Returns:
    DataFrame: Modified DataFrame with the 'sequence id' column.
    """
    # Initialize a new column for the sequence id
    df['sequence_id'] = None

    # Iterate through the DataFrame
    for i in range(len(df)):
        # Initialize an empty list to store sequence information
        sequence_info = []
        current_batter = df.iloc[i]['batter']

        # Collect information from the previous `num_pitches` pitches for the same batter
        for j in range(1, num_pitches + 1):
            # Ensure we don't go out of bounds and the batter is the same
            if i - j >= 0 and df.iloc[i - j]['batter'] == current_batter:
                # Extract relevant information for the pitch
                pitch_info = df.iloc[i - j]
                count = str(pitch_info['balls']) + '-' + str(pitch_info['strikes'])
                pitch_type = pitch_info['pitch_type']
                result = pitch_info['description']

                # Create a string representation of the pitch information
                pitch_sequence = f"{count}_{pitch_type}_{result}"

                # Append to the sequence information list
                sequence_info.append(pitch_sequence)
            else:
                # If the at-bat has not had `num_pitches`, return 0
                sequence_info.append('0')
                break

        # Maintain chronological order of the sequence
        sequence_id_str = '|'.join(sequence_info)  # Keep the original order
        df.at[i, 'sequence_id'] = sequence_id_str

    return df

# Apply the function to create a sequence id based on at-bat history for the last three pitches
df_with_at_bat_sequence_ordered = create_sequence_id_at_bat_ordered(comb_pitchers_clean, num_pitches=3)

# Display the first few rows of the modified DataFrame
df_with_at_bat_sequence_ordered[['batter', 'sequence_id', 'balls', 'strikes', 'description']].head()
```

```python
# Reimplementing the function with the correct approach

def mark_xbh_in_future_at_bats(df):
    # Flagging rows with extra base hits for each batter-pitcher combination
    df['extra_base_hit'] = df['events'].isin(['double', 'triple', 'home_run'])

    # Initializing the XBH column
    df['XBH'] = 0

    # Iterating through each unique game, batter, and pitcher combination
    for (game, batter, pitcher), group in df.groupby(['game_date', 'batter', 'pitcher']):
        # Check if there's an extra base hit in any of the at-bats
        if group['extra_base_hit'].any():
            # If yes, mark all future at-bats in the game for this batter-pitcher combination with a 1
            future_at_bats_index = group.index[group.index > group[group['extra_base_hit']].index[0]]
            df.loc[future_at_bats_index, 'XBH'] = 1

    return df

# Applying the function to the dataframe
df_with_at_bat_sequence_ordered = mark_xbh_in_future_at_bats(df_with_at_bat_sequence_ordered)

# Displaying the first few rows to verify the new column
df_with_at_bat_sequence_ordered[['game_date', 'batter', 'pitcher', 'events', 'XBH']].head()
```

```python
def clean_sequence_id(sequence):
    # Handle specific cases like '0'
    if sequence == "0":
        return "0"

    # Split the sequence into individual events
    events = sequence.split('|')

    # Process each event
    processed_events = []
    for event in events:
        # Further split each event into its components
        components = event.split('_')

        # Check if the components list has the expected number of elements
        if len(components) >= 4:
            # Extract count, pitch type, and result
            count = components[0]
            pitch_type = components[1]
            result = components[3]

            # Combine the extracted components
            processed_event = f"{count} - {pitch_type} - {result}"
        else:
            # Handle cases where the data is not in the expected format
            processed_event = " "

        processed_events.append(processed_event)

    # Join the processed events back into a single string
    return ' | '.join(processed_events)
```

```python
pitch_mapper = {'FB': ['FF', 'SI', 'FC'],
                'SL': ['SL', 'ST'],
                'CH': ['CH', 'FS', 'FO'],
                'CB': ['CU', 'KC', 'SV', 'CS']
               }
```

```python
inverse_pitch_mapper = {abbrev: key for key, abbrevs in pitch_mapper.items() for abbrev in abbrevs}

def clean_sequence_id(sequence):
    if sequence == "0":
        return "0"

    events = sequence.split('|')
    processed_events = []
    for event in events:
        components = event.split('_')
        if len(components) >= 3:
            count = components[0]
            pitch_type = components[1]  # Extract the pitch type

            # Replace the pitch type with its category from the inverse_pitch_mapper
            pitch_category = inverse_pitch_mapper.get(pitch_type, pitch_type)

            result = components[2]
            processed_event = f"{count} - {pitch_category} - {result}"
        else:
            processed_event = " "

        processed_events.append(processed_event)

    return ' | '.join(processed_events)

# Apply the function to the sequence_id column
df_with_at_bat_sequence_ordered['clean_sequence_id'] = df_with_at_bat_sequence_ordered['sequence_id'].apply(clean_sequence_id)

# Check the modified DataFrame
df_with_at_bat_sequence_ordered.head(25)
```
